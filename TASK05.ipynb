```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x = 'Survived',data = df)
plt.title('Survival Count: 0 = Did not Survive, 1 = Survived')
plt.show()
```

```python
correlation = df.corr()

plt.figure(figsize=(8,6))
sns.heatmap(correlation,annot = True, cmap = 'coolwarm', fmt = '.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df['Survived'],bins=2,kde=True,color='skyblue')
plt.title('Survival Distribution (0 = Did not Survive, 1 = Survived)')
plt.xlabel('Survived')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Survived', data = df)
plt.title('Boxplot of Survival')
plt.show()
```

## OBSERVATIONS:
### Observations for Countplot of 'Survived':
- The countplot reveals that there are more **non-survivors (0)** than **survivors (1)** in the dataset. This suggests that most passengers did not survive.

### Observations for Heatmap of 'Survived' vs 'PassengerId':
- The heatmap indicates no significant correlation between **'Survived'** and **'PassengerId'**. Since **'PassengerId'** is a unique identifier for each passenger, this is expected.

### Observations for Histogram of 'Survived':
- The histogram shows the distribution of **survivors and non-survivors**, where the non-survivors have a higher frequency, reinforcing the countplot's observation.

### Observations for Boxplot of 'Survived':
- A boxplot for **'Survived'** would likely show two distinct categories: **0 (non-survivors)** and **1 (survivors)**, and since there’s no numeric data for survival, this plot would mostly reinforce the count of survivors vs. non-survivors.


## Summary of Findings:
- The Titanic dataset shows a **significant imbalance**, with **more non-survivors** than survivors.
- **'PassengerId'** has no impact on survival as it's just a unique identifier, which aligns with the lack of correlation in the heatmap.
- Further analysis with more relevant features like **age, gender, class, or embarked** would provide more insights into the factors affecting survival rates.
